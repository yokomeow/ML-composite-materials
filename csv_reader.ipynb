```python
lambdas = []
x, y, z, v = [], [], [], []

for file in glob.glob('*.csv'):
    
    # считываем csv-файл целиком
    data = pd.read_csv(file, 
                       names=['t', 'vx', 'vy', 'vz', 'sxx', 'sxy', 'sxz', 'syy', 'syz', 'szz'],
                       sep=',')
    
    # запоминаем данные только компонент скорости
    x.append(data['vx'])
    y.append(data['vy'])
    z.append(data['vz'])
    
    # считаем модуль скорости
    v.append(np.sqrt(data['vx']**2 + data['vy']**2 + data['vz']**2))
    
    # вытаскиваем из названия файла значение лямбды  (lambda=4242,42.csv)
    lam = float(file.split('=')[-1].split('.')[0].replace(',', '.'))
    
    # сохраняем лямбды для каждого файла
    lambdas.append(lam)
```
